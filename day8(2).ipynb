```python
# replacing length - i-1  with length - (i-1)

def reverse(li):
    length = len (li)
    for i in range (length//2):
        li[i] , li[length - (i-1)] = li[length - (i-1)] , li[i]
        
li=[1,2,3,4,5,6,7,8]
reverse(li)
print(li)
```

Reverse indexing concept can also be used here - 

```python
def reverse(li):
    length = len (li)
    for i in range (length//2):
        li[i] , li[ - i-1] = li[- i-1] , li[i]
        
li=[1,2,3,4,5,6,7,8]
reverse(li)
print(li)
```
