##  Functions

```python
n = int(input()) 
r = int(input())

n_fact = 1
for i in range (1, n+1):
    n_fact = n_fact*i

r_fact = 1
for i in range (1, r+1):
    r_fact = r_fact*i

# for n-r
n_r_fact = 1
for i in range (1, n-r +1):
    n_r_fact = n_r_fact*i
    
ans = n_fact//(r_fact*n_r_fact ) 
print(ans)
```

```python
def factorial(n):
    n_fact = 1
    for i in range (1, n+1):
        n_fact = n_fact*i
    return n_fact
```

```python
a= factorial(99)
a
```

```python
def fact(n):
    n_fact = 1
    for i in range (1, n+1):
        n_fact = n_fact*i
    return n_fact
def ncr(n,r):
    n_fact=fact(n)
    r_fact=fact(r)
    n_r_fact=fact(n-r)
    x = n_fact//(r_fact*n_r_fact)
    return x
a= ncr(4,3)
print(a)
```
