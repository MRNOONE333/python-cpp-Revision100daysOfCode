```python
li2=[1,2,3,4,5,6,7,8,9]
even_c_li_noCondition = [x**3 for x in li2   ]
print(even_c_li_noCondition)
even_c_li_2 = [x**3 for x in li2 if x%2 ==0  ]
print(even_c_li_2)
even_c_li_with2Conditionals = [x**3 for x in li2 if x%2 ==0 if x%3 ==0 ]
print(even_c_li_with2Conditionals)
```

```python
# if i want to print common elements.

li_1 = [1,2,3,4,5]
li_2 = [2,4,6,7]

li_intersection =[]
for ele1 in li_1 :
    for ele2 in li_2 :
        if ele1 == ele2:
            li_intersection.append(ele1)

print(li_intersection)
```
