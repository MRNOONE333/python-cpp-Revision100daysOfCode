### Linear search

```python
n = int(input())


def search(li ,n):
    i=0
    while i<len(li):
        if n == li[i]:
            return True
        i= i+1
    return False

li = [1,2,3,4,5]


if search (li , n):
    print('found')
else:
    print('not found')
```

## Binary Search



case 2 if cant element not in array.



```python
y = int(input())

arr = [1,2,3,4,5,6]
def BinarySearch(arr , ele):
    start = 0
    end = len(arr) -1
    
    while start <= end :
        mid = (start+end)// 2
        if (arr[mid] == ele) :
            return mid
        elif (arr[mid]< ele):
            start = mid +1
        else :
            end = mid -1   
    return -1


index = BinarySearch(arr , y)
print(index)
```

## Selection Sort.  


```python
# Selection sort 
def SelectionSort(arr):
    length = len(arr) 
    for i in range (length-1 ):
        MinIndex = i 
        for j in range (i+1 , length):
            if (arr[j]< arr[MinIndex]):
                MinIndex = j
        arr[i] , arr[MinIndex] = arr[MinIndex] , arr[i]
    
arr = [1,2,100,9,5,1,2,100,9,51,2,100,9,51,2,100,9,51,2,100,9,51,2,100,9,51,2,100,9,51,2,100,9,51,2,100,9,5]
SelectionSort(arr)
print(arr)
```

```python
# Using library/Imports

start_time = time.time()

SelectionSort(arr)

# Get the current time after sorting
end_time = time.time()

# Calculate the execution time
execution_time = end_time - start_time

print(execution_time)
```
