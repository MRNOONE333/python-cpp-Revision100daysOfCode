```python
n = len (li)
m = len(li[0])


def fuction_largest_sum(li):
    max_sum=-1
    max_index=-1
    for j in range(m) :
        sum = 0 
        for ele in li :
            # can skip the  for every row . (helps to reduce the space complexity.)
            sum+= ele[j]
            if sum> max_sum:
                max_sum = sum
                max_index =j
    return max_sum , max_index

li=[[1,2,3],[4,5,6] , [7,8,9]]

largest_column_sum , largest_index =  fuction_largest_sum(li)
print( largest_column_sum , largest_index )
```
